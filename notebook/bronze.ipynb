```python
import kagglehub
from kagglehub import KaggleDatasetAdapter
```

```python
import shutil
from pathlib import Path
```

```python
import pandas as pd
```

```python
# Download latest version
download_path = kagglehub.dataset_download("samuelcortinhas/gdp-of-european-countries")
```

```python
source_path = Path(download_path)
```

```python
datalake_path = Path("./datalake/")
```

```python
datalake_path.mkdir(parents=True, exist_ok=True)
```

```python
for file_path in source_path.iterdir():
    if file_path.is_file():
        destination = datalake_path / file_path.name
        
        shutil.move(str(file_path), str(destination))
        print(f"Moved: {file_path.name}")
```

```python
dtype = {
    "year": "Int64",
    "Belgium": "Int64",
    "France": "Int64",
    "Germany": "Int64",
    "Italy": "Int64",
    "Poland": "Int64",
    "Spain": "Int64"
}
```

```python
df = pd.read_csv(
    datalake_path / 'GDP_table.csv',
    dtype=dtype
)
```

```python
# Display first rows
df.head()
```

```python
from sqlalchemy import create_engine
engine = create_engine('postgresql+psycopg://root:root@localhost:5432/eu_gdp')
```

```python
print(pd.io.sql.get_schema(df, name='eu_gdp', con=engine))
```

```python
df.head(n=0).to_sql(name='eu_gdp', con=engine, if_exists='replace')
```
